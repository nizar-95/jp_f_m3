```python
import PyPDF2
import re
import spacy
```

```python


# Open the PDF file
pdf_file = open('Chan2014a.pdf', 'rb')

# Create a PDF reader object
pdf_reader = PyPDF2.PdfReader(pdf_file)

# Extract text from the PDF file
text = ''

for page_num in range(len(pdf_reader.pages)):
    page = pdf_reader.pages[page_num]
    text += page.extract_text()
    print(text)

pdf_file.close()
```

```python
# references = re.findall(r'\bReferences?\b.*\n((?:.+\n)+)', text)
references = re.findall(r'\n(.+?\(\d{4}\)[^()\n]*(?:\([^()\n]*\)[^()\n]*)*(?:pp\d+[-\d+]*)?)\n', text)
print(references)
```

```python
# check similarities using method2
nlp_md = spacy.load("en_core_web_md")
doc_md = nlp_md(text)
ref_docs = [nlp_md(ref) for ref in references]
# Compute the similarity between the text and each reference
similarity_scores = [doc_md.similarity(ref) for ref in ref_docs]
#  Print the similarity scores
for i, score in enumerate(similarity_scores):
    print(f"Reference {i+1}: {score}")
```
